```python
## To get Position of the Outlier (i.e. Indices)
print(np.where(df['Age']>66))
age_ind = np.where(df['Age']>66)
print(type(age_ind))
print(len(age_ind[0]))
```

```python
### We may Drop rows having outliers for 'Age' column
print("Before", df.shape)
df.drop(age_ind[0]) ### Use inplace= True if need be
print("After ", df.shape)
```

```python
q1 = np.quantile(df['Age'], 0.25)
q3 = np.quantile(df['Age'], 0.75)
IQR = q3-q1
IQR
```

```python
### Use IQR to find lower and upper limit
lower_limit = q1 - 1.5*IQR
upper_limit = q3 + 1.5*IQR
print(lower_limit,upper_limit)
```

```python
### Use these limits to remove instances (rows)
index_list = df.index[ (df['Age'] < lower_limit) | (df['Age'] > upper_limit) ]
print(index_list)
df = df.drop(index_list)
```

```python
df.shape
```

```python
### Finding the Boundary Values for Outliers
print("For Age")
upper_limit = df['Age'].mean() + 3*df['Age'].std()
lower_limit = df['Age'].mean() - 3*df['Age'].std()
print("Highest Potential Age of Diabetes",upper_limit)
print("Lowest Potential Age of Diabetes",lower_limit)
print(upper_limit,lower_limit)
## Applying the limits
df["Age"] = np.where (df["Age"] <lower_limit, lower_limit,df['Age'])
df["Age"] = np.where(df["Age"] >upper_limit, upper_limit,df['Age'])
```
